### Ï£ºÍ∞Ä ÌÉêÏÉâÏ†Å Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù(EDA, Exploratory Data Analysis)

Ï£ºÏãù ÏãúÏû• Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÏÑùÌïòÏó¨, Îç∞Ïù¥ÌÑ∞Ïùò ÌäπÏÑ±ÏùÑ Ïù¥Ìï¥ÌïòÍ≥†, Ï§ëÏöîÌïú Ìå®ÌÑ¥, Ìä∏Î†åÎìú, Í¥ÄÍ≥Ñ Îì±ÏùÑ ÌååÏïÖÌïòÎäî Í≥ºÏ†ïÏûÖÎãàÎã§.

### EDAÏùò Ï£ºÏöî Î™©Ï†Å

1. **Îç∞Ïù¥ÌÑ∞Ïùò Í∏∞Ï¥àÏ†ÅÏù∏ ÌäπÏÑ± Ïù¥Ìï¥**:

3. **ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ Î∂ÑÏÑù**:
   - Ïó¨Îü¨ Ï£ºÏãù Ï¢ÖÎ™© Í∞Ñ, ÌòπÏùÄ Ï£ºÏãù Í∞ÄÍ≤©Í≥º Îã§Î•∏ Í≤ΩÏ†ú ÏßÄÌëú(Í∏àÎ¶¨, ÌôòÏú® Îì±) Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º Î∂ÑÏÑùÌï©ÎãàÎã§.
   - ÏòàÎ•º Îì§Ïñ¥, Ï£ºÏãùÍ≥º Í∏àÎ¶¨Ïùò ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÎ•º Î∂ÑÏÑùÌïòÏó¨ Í∏àÎ¶¨Í∞Ä ÏÉÅÏäπÌï† Îïå Ï£ºÏãù ÏãúÏû•Ïù¥ Ïñ¥ÎñªÍ≤å Î∞òÏùëÌïòÎäîÏßÄ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

4. **Ïù¥ÏÉÅÏπò Î∞è Í≤∞Ï∏°Ïπò Î∂ÑÏÑù**:
   - Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïù¥ÏÉÅÏπòÎÇò Í≤∞Ï∏°ÏπòÎ•º Ï∞æÏïÑÎÇ¥Ïñ¥ Îç∞Ïù¥ÌÑ∞ ÌíàÏßàÏùÑ ÌèâÍ∞ÄÌïòÍ≥†, Ïù¥Î•º Ïñ¥ÎñªÍ≤å Ï≤òÎ¶¨Ìï†ÏßÄ Í≤∞Ï†ïÌï©ÎãàÎã§.
   - ÏòàÎ•º Îì§Ïñ¥, Í∞ëÏûëÏä§Îü¨Ïö¥ Í∏âÎì±Ïù¥ÎÇò Í∏âÎùΩÏù¥ Ïù¥ÏÉÅÏπòÏùº Ïàò ÏûàÏäµÎãàÎã§.

5. **ÏãúÍ∞ÅÌôî**:
   - Ï£ºÍ∞ÄÏùò Î≥ÄÎèôÏÑ±ÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌëúÌòÑÌïòÏó¨ ÏâΩÍ≤å Ïù¥Ìï¥Ìï† Ïàò ÏûàÎèÑÎ°ù Ìï©ÎãàÎã§. ÏãúÍ≥ÑÏó¥ Í∑∏ÎûòÌîÑ, ÌûàÏä§ÌÜ†Í∑∏Îû®, Î∞ïÏä§ÌîåÎ°Ø, ÏÉÅÍ¥Ä ÌñâÎ†¨ Îì±Ïù¥ ÏûêÏ£º ÏÇ¨Ïö©Îê©ÎãàÎã§.
   - ÏòàÎ•º Îì§Ïñ¥, Ï£ºÍ∞ÄÏùò Î≥ÄÌôî Ï∂îÏù¥Î•º ÏÑ† Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôîÌïòÍ±∞ÎÇò, Ï£ºÏãù Í∞Ñ ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÎ•º ÌûàÌä∏ÎßµÏúºÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÏäµÎãàÎã§.

   - ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÌëúÏ§ÄÌé∏Ï∞®, ÏµúÏÜåÍ∞í, ÏµúÎåÄÍ∞í Îì±ÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Ïùò Í∏∞Ï¥àÏ†ÅÏù∏ ÌäπÏÑ±ÏùÑ ÌååÏïÖÌï©ÎãàÎã§.

### Ï£ºÍ∞Ä EDA Î∂ÑÏÑù ÌùêÎ¶Ñ
#### 1. üì• Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞

```python
import yfinance as yf

df = yf.download("005930.KS", start="2024-01-01", end="2024-12-31")  # ÏÇºÏÑ±Ï†ÑÏûê
print(df)
print('----------------droplevel multi index columnÏàòÏ†ï----------------')
df.columns = df.columns.droplevel(1)  # Îëê Î≤àÏß∏ Î†àÎ≤® Ï†úÍ±∞
print(df)
df.tail()
```

#### 2. üßπ Í∏∞Ï¥à Ï†ïÏ†ú

```python
df = df.dropna()  # Í≤∞Ï∏°Ïπò Ï†úÍ±∞
print(df)
# df.index = pd.to_datetime(df.index)  # Ïù∏Îç±Ïä§Î•º ÎÇ†ÏßúÎ°ú
```

#### 3. üìà ÏãúÍ∞ÅÏ†Å ÌôïÏù∏ (Ï¢ÖÍ∞Ä, Í±∞ÎûòÎüâ)

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÏùåÏàò Î∂ÄÌò∏ Íπ®Ïßê Î∞©ÏßÄ

plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close Price')
plt.title('Ï¢ÖÍ∞Ä Ï∂îÏù¥')
plt.legend()
plt.show()

df['Volume'].plot(kind='bar', figsize=(12, 4), title='Í±∞ÎûòÎüâ', color='grey')
plt.tight_layout()
plt.show()
```

#### 4. üìä Í∏∞Î≥∏ ÌÜµÍ≥Ñ Î∂ÑÏÑù

```python
df.describe()  # ÌèâÍ∑†, ÌëúÏ§ÄÌé∏Ï∞®, ÏµúÏÜå/ÏµúÎåÄÍ∞í Îì±
```

#### 5. üåÄ Ïù¥ÎèôÌèâÍ∑†ÏÑ†(MA) Ï∂îÍ∞Ä

```python
df['MA5'] = df['Close'].rolling(window=5).mean()
df['MA20'] = df['Close'].rolling(window=20).mean()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close')
plt.plot(df['MA5'], label='MA5')
plt.plot(df['MA20'], label='MA20')
plt.title('Ïù¥ÎèôÌèâÍ∑†ÏÑ† ÎπÑÍµê')
plt.legend()
plt.show()
```

#### 6. üìâ ÏàòÏùµÎ•† Í≥ÑÏÇ∞

```python
df['Daily Return'] = df['Close'].pct_change()
df['Cumulative Return'] = (1 + df['Daily Return']).cumprod()
```

### Í∑∏ÎûòÌîÑÍ∞Ä Î≥¥Ïó¨Ï£ºÎäî Í≤É

Í∞ÄÏû• ÏûêÏ£º ÎÇòÏò§Îäî ÏàòÏùµÎ•† Íµ¨Í∞ÑÏù¥ Ïñ¥ÎîîÏù∏ÏßÄ ÌååÏïÖ

```python
plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="hist", bins=100, title="ÏùºÎ≥Ñ ÏàòÏùµÎ•† Î∂ÑÌè¨(ÌûàÏä§ÌÜ†Í∑∏Îû®)" )  # ÌûàÏä§ÌÜ†Í∑∏Îû® Íµ¨Í∞Ñ100ÏúºÎ°ú ÎÇòÎàî
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="bar", title="ÏùºÎ≥Ñ ÏàòÏùµÎ•† Î∂ÑÌè¨(Î∞î)")
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="bar", title="ÏùºÎ≥Ñ ÏàòÏùµÎ•† Î∂ÑÌè¨(Î∞î, xÏ∂ï ÎààÍ∏à Ï†úÍ±∞)").set_xticks([])
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="kde", title="ÏùºÎ≥Ñ ÏàòÏùµÎ•† Î∂ÑÌè¨(Ïª§ÎÑêÎ∞ÄÎèÑÏ∂îÏ†ïÍ∑∏ÎûòÌîÑÌîÑ)")
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(title="ÎàÑÏ†Å ÏàòÏùµÎ•†")
plt.show()

plt.figure(figsize=(12, 4))
df['Cumulative Return'].plot(title='ÎàÑÏ†Å ÏàòÏùµÎ•†(xÏ∂ï ÎùºÎ≤®Ï†úÍ±∞)').set_xticklabels([])  # xÏ∂ï ÎùºÎ≤® Ï†úÍ±∞
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(title="ÎàÑÏ†Å ÏàòÏùµÎ•†(xÏ∂ï ÎààÍ∏àÏ†úÍ±∞)").set_xticks([])  # xÏ∂ï ÎààÍ∏à ÎùºÎ≤® Ï†úÍ±∞
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(kind='kde', title="ÎàÑÏ†Å ÏàòÏùµÎ•†(xÏ∂ï ÎààÍ∏àÏ†úÍ±∞)")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÏùåÏàò Î∂ÄÌò∏ Íπ®Ïßê Î∞©ÏßÄ


# ÏÇºÏÑ±Ï†ÑÏûê Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú
df = yf.download('005930.KS', start='2024-01-01', end='2025-01-01')

print(df)
print('----------------droplevel multi index columnÏàòÏ†ï----------------')
df.columns = df.columns.droplevel(1)  # Îëê Î≤àÏß∏ Î†àÎ≤® Ï†úÍ±∞
print(df)

# ÏµúÎåÄ Ï¢ÖÍ∞Ä Î∞è Ìï¥Îãπ ÎÇ†Ïßú Íµ¨ÌïòÍ∏∞
max_close = df['Close'].max()
max_date = df['Close'].idxmax()

print(f"ÏµúÎåÄ Ï¢ÖÍ∞Ä: {max_close:.2f}")
print(f"Ìï¥Îãπ ÎÇ†Ïßú: {max_date.date()}")


# ÏµúÏÜå Ï¢ÖÍ∞Ä Î∞è Ìï¥Îãπ ÎÇ†Ïßú Íµ¨ÌïòÍ∏∞
min_close = df['Close'].min()
min_date = df['Close'].idxmin()

print(f"ÏµúÏÜå Ï¢ÖÍ∞Ä: {min_close:.2f}")
print(f"Ìï¥Îãπ ÎÇ†Ïßú: {min_date.date()}")

# Ï¢ÖÍ∞Ä Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Ï¢ÖÍ∞Ä', color='blue')

# ÏµúÎåÄÍ∞íÏóê ÏõêÌòï ÎßàÏª§ ÌëúÏãú
plt.plot(max_date, max_close, 'ro', label='ÏµúÎåÄ Ï¢ÖÍ∞Ä')  # Îπ®Í∞Ñ ÎèôÍ∑∏ÎùºÎØ∏

# ÏµúÏÜåÍ∞íÏóê ÏõêÌòï ÎßàÏª§ ÌëúÏãú
plt.plot(min_date, min_close, 'bo', label='ÏµúÏÜå Ï¢ÖÍ∞Ä')  # ÌååÎûÄ ÎèôÍ∑∏ÎùºÎØ∏

# ÏµúÎåÄ~ÏµúÏÜå ÏÇ¨Ïù¥ ÏÑ† Í∏ãÍ∏∞
plt.plot([min_date, max_date], [min_close, max_close], 'k--', label='ÏµúÎåÄ-ÏµúÏÜå Ïó∞Í≤∞ÏÑ†')  # Í≤ÄÏ†ï Ï†êÏÑ†


# ÌÖçÏä§Ìä∏ Ï£ºÏÑù
plt.text(max_date, max_close, f'{max_close:.2f}', color='red', fontsize=10, ha='left', va='bottom')
plt.text(min_date, min_close, f'{min_close:.2f}', color='blue', fontsize=10, ha='right', va='top')



# Íæ∏ÎØ∏Í∏∞
plt.title('ÏÇºÏÑ±Ï†ÑÏûê Ï¢ÖÍ∞Ä Í∑∏ÎûòÌîÑ')
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Ï¢ÖÍ∞Ä')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Ï∂îÏ∂úÎç∞Ïù¥ÌÑ∞ Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞

# Ï∂îÏ∂ú Ï¢ÖÍ∞Ä Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

# ÏµúÎåÄ/ÏµúÏÜå Ï¢ÖÍ∞ÄÏóê Ìï¥ÎãπÌïòÎäî ÌñâÎßå Ï∂îÏ∂ú
# select_df = df.loc[[min_date, max_date]]  # ÎëêÎÇ†ÏûêÏùò Í∞íÎßå Ï∂îÏ∂ú
# ÎÇ†Ïßú ÏàúÏÑú Ï†ïÎ¶¨ (ÎßåÏïΩ ÏµúÏÜåÍ∞Ä Îí§Ïóê ÏûàÎã§Î©¥ ÏàúÏÑú Î∞îÍøîÏ§å)

start_date = min(min_date, max_date)
end_date = max(min_date, max_date)

select_df = df.loc[start_date : end_date]  # ÎëêÎÇ†Ïûê ÏÇ¨Ïù¥Ïùò Î≤îÏúÑ ÏßÄÏ†ï
print('-----------select df -------------------')
print(select_df)

plt.figure(figsize=(5,6))
plt.plot(select_df['Close'], label='Ï¢ÖÍ∞Ä', color='yellow')

plt.title('ÏÇºÏÑ±Ï†ÑÏûê ÏÑ†ÌÉù Ï¢ÖÍ∞Ä Í∑∏ÎûòÌîÑ')
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Ï¢ÖÍ∞Ä')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 25% Î∂ÑÏúÑÍ∞íÍ≥º Ìï¥Îãπ ÎÇ†Ïßú Ï∞æÍ∏∞

```python
# Ï¢ÖÍ∞ÄÏùò 25% Î∂ÑÏúÑÏàò Í≥ÑÏÇ∞
q25 = df['Close'].describe()['25%']

# 25% Î∂ÑÏúÑÏàòÏôÄ Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï¢ÖÍ∞Ä Ï∞æÍ∏∞
closest_idx = (df['Close'] - q25).abs().idxmin()
closest_value = df.loc[closest_idx, 'Close']

print(f"25% Î∂ÑÏúÑÍ∞í: {q25:.2f}")
print(f"Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÎÇ†Ïßú: {closest_idx.date()}, Ï¢ÖÍ∞Ä: {closest_value:.2f}")
```
