## 3. 탐색적 데이터 분석(EDA)

### 주가 탐색적 데이터 분석(EDA, Exploratory Data Analysis)

주식 시장 데이터를 분석하여, 데이터의 특성을 이해하고, 중요한 패턴, 트렌드, 관계 등을 파악하는 과정입니다.  
EDA는 주가 데이터를 깊이 있게 분석하여 예측 모델을 구축하거나 투자 결정을 지원할 수 있는 중요한 첫걸음입니다.

### EDA의 주요 목적

1. **데이터의 기초적인 특성 이해**:
   - 주가 데이터의 분포, 평균, 표준편차 등 기초 통계량을 계산하여 데이터의 전반적인 특성을 파악합니다.
   - 예를 들어, 주식의 가격이 상승하는 경향이 있는지, 하락하는 경향이 있는지를 파악할 수 있습니다.

2. **데이터의 시계열 특성 분석**:
   - 주가는 시간이 지남에 따라 변화하는 시계열 데이터이므로, 시간에 따른 패턴이나 주기성을 파악하는 것이 중요합니다.
   - 예를 들어, 주가의 이동평균(Moving Average), Bollinger Bands, 또는 계절성과 같은 패턴을 찾아낼 수 있습니다.

3. **상관 관계 분석**:
   - 여러 주식 종목 간, 혹은 주식 가격과 다른 경제 지표(금리, 환율 등) 간의 관계를 분석합니다.
   - 예를 들어, 주식과 금리의 상관 관계를 분석하여 금리가 상승할 때 주식 시장이 어떻게 반응하는지 확인할 수 있습니다.

4. **이상치 및 결측치 분석**:
   - 주가 데이터에서 이상치나 결측치를 찾아내어 데이터 품질을 평가하고, 이를 어떻게 처리할지 결정합니다.
   - 예를 들어, 갑작스러운 급등이나 급락이 이상치일 수 있습니다.

5. **시각화**:
   - 주가의 변동성을 시각적으로 표현하여 쉽게 이해할 수 있도록 합니다. 시계열 그래프, 히스토그램, 박스플롯, 상관 행렬 등이 자주 사용됩니다.
   - 예를 들어, 주가의 변화 추이를 선 그래프로 시각화하거나, 주식 간 상관 관계를 히트맵으로 나타낼 수 있습니다.

### EDA의 과정

1. **데이터 준비**:
   - 주가 데이터를 수집하고, 필요한 컬럼(예: 날짜, 시가, 종가, 고가, 저가, 거래량 등)을 준비합니다.
   - 결측치나 이상치를 확인하고, 이를 처리합니다.

2. **기초 통계 분석**:
   - 평균, 중앙값, 표준편차, 최소값, 최대값 등을 계산하여 데이터의 기초적인 특성을 파악합니다.

3. **시각화**:
   - 시계열 그래프, 히스토그램, 상관 행렬 등을 활용하여 데이터를 시각적으로 분석합니다.

4. **패턴 및 트렌드 발견**:
   - 주가의 주요 패턴(예: 상승, 하락, 변동성)을 식별하고, 이를 기반으로 향후 예측 모델이나 투자 전략을 수립할 수 있습니다.

주가 EDA는 데이터 분석의 첫 단계로, 데이터에 대한 직관적이고 철저한 이해를 돕고, 모델링 과정에 앞서 데이터를 준비하는 중요한 역할을 합니다.

## 용어 정리(퀀트전략)

**퀀트 전략(Quantitative Strategy)**은 주관적인 판단 대신 수학적·통계적 모델, 데이터 분석, 알고리즘을 기반으로 자산을 매매하는 투자 전략입니다.

"Quant"는 Quantitative = 정량적이라는 뜻에서 나온 말이에요.

✅ 핵심 개념
| 항목                | 설명                             |
| ----------------- | ------------------------------ |
| 📊 **데이터 기반**     | 가격, 거래량, 재무지표 등 숫자 데이터를 바탕으로 함 |
| 🧠 **알고리즘/모델 사용** | 수학적 규칙이나 통계적 분석, 머신러닝 모델 등을 활용 |
| ⚙️ **자동화 가능**     | 전략을 프로그래밍하여 반복적으로 매매 가능        |
| ❌ **감정 배제**       | 시장 뉴스나 감정에 휘둘리지 않음             |


✅ 퀀트 전략의 예시  
| 전략 종류        | 설명                          |
| ------------ | --------------------------- |
| **모멘텀 전략**   | 최근 수익률이 높은 종목에 투자           |
| **가치 투자 전략** | 저PER, 저PBR 같은 지표 기반         |
| **평균회귀 전략**  | 가격이 평균에서 멀어지면 다시 돌아온다는 전제   |
| **페어 트레이딩**  | 상관관계 높은 두 자산의 가격 차이를 이용한 매매 |
| **시장중립 전략**  | 수익은 추구하되 전체 시장 방향과는 무관하게 설정 |


✅ 퀀트 전략 개발 과정
아이디어 구상 (예: 20일선 돌파 매매)

데이터 수집 (예: yfinance, Quandl 등)

전략 백테스트 (과거 데이터로 성능 검증)

리스크 조정 및 최적화

실전 매매 적용 (알고리즘 트레이딩)

✅ 장점과 단점
| 장점           | 단점                   |
| ------------ | -------------------- |
| 일관성 있고 감정 없음 | 시장 급변동엔 약할 수 있음      |
| 자동화 가능       | 과최적화(overfitting) 주의 |
| 다양한 전략 조합 가능 | 데이터 품질과 처리 중요        |


요약
퀀트 전략은 객관적 데이터와 수학 모델에 기반하여 사람이 직접 판단하지 않아도 작동하는 투자 전략입니다.



### 주가 EDA 분석 흐름
#### 1. 📥 데이터 불러오기

```python
import yfinance as yf

df = yf.download("005930.KS", start="2024-01-01", end="2024-12-31")  # 삼성전자
print(df)
print('----------------droplevel multi index column수정----------------')
df.columns = df.columns.droplevel(1)  # 두 번째 레벨 제거
print(df)
df.tail()
```

#### 2. 🧹 기초 정제

```python
df = df.dropna()  # 결측치 제거
print(df)
# df.index = pd.to_datetime(df.index)  # 인덱스를 날짜로
```

#### 3. 📈 시각적 확인 (종가, 거래량)

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 음수 부호 깨짐 방지

plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close Price')
plt.title('종가 추이')
plt.legend()
plt.show()

df['Volume'].plot(kind='bar', figsize=(12, 4), title='거래량', color='grey')
plt.tight_layout()
plt.show()
```

#### 4. 📊 기본 통계 분석

```python
df.describe()  # 평균, 표준편차, 최소/최대값 등
```

| 항목           | 의미                   |
| ------------ | -------------------- |
| `mean`       | 평균 수익률               |
| `std`        | 수익률의 변동성             |
| `min`, `max` | 최대 낙폭, 최대 상승률        |
| `25%`, `75%` | 분위수 → 대략적인 수익률 범위 파악 |


#### 5. 🌀 이동평균선(MA) 추가

```python
df['MA5'] = df['Close'].rolling(window=5).mean()
df['MA20'] = df['Close'].rolling(window=20).mean()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close')
plt.plot(df['MA5'], label='MA5')
plt.plot(df['MA20'], label='MA20')
plt.title('이동평균선 비교')
plt.legend()
plt.show()
```

#### 6. 📉 수익률 계산

```python
df['Daily Return'] = df['Close'].pct_change()
df['Cumulative Return'] = (1 + df['Daily Return']).cumprod()
```

### 그래프가 보여주는 것

가장 자주 나오는 수익률 구간이 어디인지 파악

양(+)의 수익 vs 음(–)의 수익 비중

극단적인 변동 (이상치) 도 시각적으로 탐지 가능

```python
plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="hist", bins=100, title="일별 수익률 분포(히스토그램)" )  # 히스토그램 구간100으로 나눔
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="bar", title="일별 수익률 분포(바)")
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="bar", title="일별 수익률 분포(바, x축 눈금 제거)").set_xticks([])
plt.show()

plt.figure(figsize=(12, 4))
df["Daily Return"].plot(kind="kde", title="일별 수익률 분포(커널밀도추정그래프프)")
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(title="누적 수익률")
plt.show()

plt.figure(figsize=(12, 4))
df['Cumulative Return'].plot(title='누적 수익률(x축 라벨제거)').set_xticklabels([])  # x축 라벨 제거
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(title="누적 수익률(x축 눈금제거)").set_xticks([])  # x축 눈금 라벨 제거
plt.show()

plt.figure(figsize=(12, 4))
df["Cumulative Return"].plot(kind='kde', title="누적 수익률(x축 눈금제거)")
plt.show()
```

#### 7. 📌 변동성 확인 (표준편차)

```python
volatility = df['Daily Return'].rolling(window=20).std()
plt.figure(figsize=(12, 4))
volatility.plot(title='20일 이동 변동성')
plt.show()
```

### 변동성을 확인하는 이유


| 이유                   | 설명                                                            |
| -------------------- | ------------------------------------------------------------- |
| 📉 **리스크(위험) 측정**    | 변동성이 크면 가격이 불안정하고 리스크가 높다는 뜻입니다. 안정적인 종목을 찾거나 리스크를 회피할 수 있어요. |
| 🧠 **시장 심리 파악**      | 시장이 불안할 때(예: 위기, 급락 직전) 변동성이 급등합니다. 투자자 심리를 간접적으로 알 수 있어요.    |
| 🎯 **전략 조정**         | 변동성이 높을 땐 손절 라인을 넓히고, 낮을 땐 타이트하게 설정하는 등 전략 조정에 활용됩니다.         |
| 📊 **옵션·파생상품 가격 결정** | 옵션 가격은 변동성을 기반으로 결정됩니다. 특히 내재변동성은 핵심 변수입니다.                   |
| 🔍 **시장 타이밍**        | 변동성이 매우 낮아지면 큰 움직임이 올 가능성도 있어 타이밍 판단에 참고합니다.                  |


"변동성은 위험을 수치로 표현한 것이며, 퀀트 전략의 핵심 요소 중 하나"입니다.
변동성이 커지는 시점은 주의 또는 기회일 수 있어요.

#### 8. ⚠️ 이상치(Outlier) 탐색

```python
# Boxplot
import seaborn as sns
print(df['Daily Return'])
sns.boxplot(df['Daily Return'])
plt.title('일별 수익률 이상치 탐지')
plt.show()
```

   이상치 (●)     ← 여기!
       |
  ─────┐ ┌─────  ← 상자 경계 밖 수염 (whisker)
       │ │
       ▼ ▼
  |───────────|  ← 상자(Box) = IQR 범위
  Q1        Q3
    └───┬───┘
        │
     중앙값 (Median)


이상치 기준: IQR 방식

Q1 (1사분위수) = 하위 25%

Q3 (3사분위수) = 상위 75%

IQR = Q3 - Q1

이상치 정의:

Q1 - 1.5×IQR 미만, 또는

Q3 + 1.5×IQR 초과하는 값들

→ 이 값들은 박스 밖에 ● 점으로 찍혀 나옵니다.
즉, 수익률이 다른 날보다 유난히 크거나 작았던 날이 이상치로 시각화돼요.

### 삼성전자 주가 데이터에서 최고종가의 날짜를 찾고 그래프에 원형마커 표시

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 음수 부호 깨짐 방지


# 삼성전자 데이터 다운로드
df = yf.download('005930.KS', start='2024-01-01', end='2025-01-01')

print(df)
print('----------------droplevel multi index column수정----------------')
df.columns = df.columns.droplevel(1)  # 두 번째 레벨 제거
print(df)

# 최대 종가 및 해당 날짜 구하기
max_close = df['Close'].max()
max_date = df['Close'].idxmax()

print(f"최대 종가: {max_close:.2f}")
print(f"해당 날짜: {max_date.date()}")


# 최소 종가 및 해당 날짜 구하기
min_close = df['Close'].min()
min_date = df['Close'].idxmin()

print(f"최소 종가: {min_close:.2f}")
print(f"해당 날짜: {min_date.date()}")

# 종가 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='종가', color='blue')

# 최대값에 원형 마커 표시
plt.plot(max_date, max_close, 'ro', label='최대 종가')  # 빨간 동그라미

# 최소값에 원형 마커 표시
plt.plot(min_date, min_close, 'bo', label='최소 종가')  # 파란 동그라미

# 최대~최소 사이 선 긋기
plt.plot([min_date, max_date], [min_close, max_close], 'k--', label='최대-최소 연결선')  # 검정 점선


# 텍스트 주석
plt.text(max_date, max_close, f'{max_close:.2f}', color='red', fontsize=10, ha='left', va='bottom')
plt.text(min_date, min_close, f'{min_close:.2f}', color='blue', fontsize=10, ha='right', va='top')



# 꾸미기
plt.title('삼성전자 종가 그래프')
plt.xlabel('날짜')
plt.ylabel('종가')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 추출데이터 그래프 꾸미기

# 추출 종가 그래프 그리기

# 최대/최소 종가에 해당하는 행만 추출
# select_df = df.loc[[min_date, max_date]]  # 두날자의 값만 추출
# 날짜 순서 정리 (만약 최소가 뒤에 있다면 순서 바꿔줌)

start_date = min(min_date, max_date)
end_date = max(min_date, max_date)

select_df = df.loc[start_date : end_date]  # 두날자 사이의 범위 지정
print('-----------select df -------------------')
print(select_df)

plt.figure(figsize=(5,6))
plt.plot(select_df['Close'], label='종가', color='yellow')

plt.title('삼성전자 선택 종가 그래프')
plt.xlabel('날짜')
plt.ylabel('종가')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 25% 분위값과 해당 날짜 찾기

```python
# 종가의 25% 분위수 계산
q25 = df['Close'].describe()['25%']

# 25% 분위수와 가장 가까운 종가 찾기
closest_idx = (df['Close'] - q25).abs().idxmin()
closest_value = df.loc[closest_idx, 'Close']

print(f"25% 분위값: {q25:.2f}")
print(f"가장 가까운 날짜: {closest_idx.date()}, 종가: {closest_value:.2f}")
```
