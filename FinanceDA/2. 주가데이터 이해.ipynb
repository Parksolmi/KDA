# **주가데이터 이해**

## 1. 주가 데이터란 무엇인가?

### 1. 주가 데이터의 정의

주가 데이터는 특정 주식(또는 지수 등)의 일정 기간 동안의 가격과 거래 정보를 기록한 데이터입니다.  
일반적으로 일, 주, 월 단위로 수집되며, 시가, 고가, 저가, 종가, 거래량, 조정 종가 등의 항목을 포함합니다.

### 2. 주가 데이터의 중요성
  - 투자 판단 기준: 투자자들이 매수/매도 타이밍을 결정할 때 근거로 사용합니다.  
  - 시장 심리 분석: 가격 변동과 거래량을 통해 시장의 매수/매도 심리를 파악할 수 있습니다.  
  - 기술적 분석의 핵심 재료: 차트 분석, 이동평균선, 지표 계산 등에 활용됩니다.  
  - 알고리즘 트레이딩/퀀트 전략 개발: 백테스트용 데이터로 쓰입니다.

### 3. 주가 데이터 활용 사례 (투자, 분석, 연구)
  - 투자: 개인 및 기관 투자자들이 종목 선정과 타이밍 판단에 사용.  
  - 분석: 애널리스트들이 기업 실적 외에도 기술적 지표를 통해 주가 예측.  
  - 연구: 금융공학, 데이터 사이언스 연구자들이 가격 예측 모델 및 전략 검증.

### 2. 주가 데이터의 기본 구성 요소

| 항목                        | 의미                                       |
| -------------------------- | ------------------------------------------ |
| **시가 (Open)**             | 하루 주식 거래가 시작될 때 최초로 체결된 가격    |
| **고가 (High)**             | 거래일 중 주식이 기록한 가장 높은 가격          |
| **저가 (Low)**              | 거래일 중 주식이 기록한 가장 낮은 가격          |
| **종가 (Close)**            | 하루 거래가 종료될 때 마지막으로 체결된 가격     |
| **거래량 (Volume)**         | 해당 주식이 거래된 총 주식 수 (매매된 양)       |
| **Adjusted Close (조정 종가)** | **배당, 액면분할, 무상증자 등** 기업 이벤트를 반영한 종가로, **실제 수익률 계산에 적합한 가격**입니다. |


## 2. 용어 정리


1. KRX (한국거래소, Korea Exchange)
   대한민국의 유일한 증권, 선물, 파생상품 거래소.

   주식(코스피, 코스닥, 코넥스), ETF, ETN, 선물·옵션 등을 매매하는 곳.

   2005년에 기존 한국증권거래소 + 한국선물거래소 + 코스닥증권시장이 통합되어 만들어짐.


2. NXT

   "NXT"는 2025년 3월 4일 한국에서 개설된 **국내 최초의 대체거래소(Alternative Trading System, ATS)**인 **넥스트레이드(NEXTRADE)**를 의미합니다.   
   이는 기존 한국거래소(KRX)의 독점 구조를 깨고, 주식 거래의 효율성과 경쟁력을 높이기 위해 도입되었습니다. 

   넥스트레이드(NXT)는 기존 KRX의 거래 시간 외에도 추가적인 거래 시간을 제공하고, 다양한 주문 유형과 낮은 수수료를 통해 투자자에게 더 많은 선택지를 제공합니다. 

   그러나 모든 증권사가 참여하는 것은 아니므로, 거래를 원하는 증권사의 참여 여부를 확인하는 것이 중요합니다.

   요약 비교: KRX vs NXT
|항목	     |한국거래소(KRX)	|  넥스트레이드(NXT)                             |
|------     |-----------------|---------------------------------------------|
|운영 시간   |	09:00 ~ 15:30	|  08:00 ~ 20:00                             |
|수수료	    |표준 수수료 적용     |	KRX 대비 20~40% 낮음                        |
|거래 종목 수|	약 2,000여 종목   |	10종목 시작, 3월 말까지 800종목 목표          |
|주문 유형   |	시장가, 지정가 주문	|시장가, 지정가, 중간가호가, 스톱지정가호가 등   |
|거래 계좌 개설|	모든 증권사 가능|	참여 증권사 한정                          |

3. 코스피 (KOSPI, Korea Composite Stock Price Index)
   대한민국 대표 종합 주가지수.

   코스피 시장에 상장된 모든 기업의 시가총액을 기준으로 만든 지수.

   흔히 "주식시장이 올랐다/내렸다" 할 때 말하는 게 코스피임.

   예) 삼성전자, 현대차, SK하이닉스 등이 포함됨.


코스피 지수 계산식

KOSPI =( 현재 시가총액/ 기준 시가총액 ) ×100

​
 
🔍 구성요소 설명
| 항목          | 설명                                      |
| ----------- | --------------------------------------- |
| **현재 시가총액** | 코스피에 포함된 모든 상장기업의 현재 시가총액 합계            |
| **기준 시가총액** | 1980년 1월 4일 코스피 시장의 시가총액 (지수를 100으로 설정) |
| **100**     | 기준 시가총액을 지수 100으로 설정한 것                 |




📌 예시

기준 시점 시가총액 = 10조 원

현재 시가총액 = 3,000조 원


30,000

KOSPI=( 3,000 /10 )×100=30,000

→ 지수가 30,000이 되었다는 뜻은 기준일 대비 시가총액이 300배 증가했다는 의미입니다.




4. 코스피200 (KOSPI 200)
코스피 시장에서 대표성 높은 200개 종목만 따로 뽑아서 만든 지수.

주로 선물, 옵션 거래의 기초자산으로 사용됨.

삼성전자, LG에너지솔루션 같은 대형주가 많이 들어있음.

코스피200 선물/옵션 거래를 통해 시장 전망에 투자할 수도 있음.



5. 코스닥 (KOSDAQ)
벤처기업이나 성장기업 중심 주식시장.

미국의 나스닥(NASDAQ) 모델을 본떠 만든 이름.

IT, 바이오, 엔터테인먼트, 2차전지 등 신산업 기업이 많음.

예) 에코프로, 셀트리온헬스케어, 카카오게임즈 등.


6. 선물 (Futures)

미래의 특정 시점에 특정 가격으로 상품이나 자산을 사거나 팔기로 약속하는 계약.

코스피200 선물, 금 선물, 원유 선물 등 존재.

투자자가 가격 상승(매수), 하락(매도)에 베팅할 수 있음. (레버리지 효과 큼)

7. 코리아밸류업 (Korea Value-up)

2024년부터 본격적으로 추진된 정책.

한국 기업들의 낮은 주가순자산비율(PBR) 문제를 해결하려는 목적.

일본의 "JPX 프라임 밸류업"처럼, 기업들이 자본 효율성을 높이도록 유도.

배당 확대, 자사주 매입, 사업구조 개선 같은 방법을 통해 기업가치(valuation)를 끌어올리려 함.

8. 코넥스 (KONEX, Korea New Exchange)

초기 소규모 기업 전용 주식시장.

코스닥보다 더 진입장벽이 낮음.

벤처, 스타트업이 상장해서 성장 후 코스닥으로 넘어가는 "사다리 시장" 역할.

거래량이 적어서 유동성 위험이 크지만, 성장 초기 기업 투자 기회도 있음.

9. ETF (Exchange Traded Fund)

주식처럼 사고파는 펀드.

지수(코스피200, S&P500 등)나 특정 테마(2차전지, 반도체)를 추종.

분산 투자 효과 + 낮은 수수료 + 편리한 매매가 장점.

(예) "TIGER 2차전지테마", "KODEX 코스피200" ETF

- 특정자산가격(국내외지수 · 상품 등)의 움직임에 수익률이 연동되도록 설계된 상품
- 거래소에 상장되며, 주식처럼 실시간 매매가 가능한 펀드
- 주식과 인덱스펀드의 장점이 결합된 상품
- 주식 : 매매의 편의성
- 인덱스펀드 : 분산투자, 낮은 거래비용

### ETF의 특징
- 환금성

실시간 매매, D+2일의 결제 주기

- 분산투자

ETF 구성 종목에 한번에 투자하는 효과

- 저투자 비용

주식형펀드, 인덱스펀드 대비 저렴한 보수율

- 운용의 투명성

ETF의 포트폴리오 내역 일일확인 가능

### ETF의 종류

- 지수추적 ETF

국내지수/해외지수를 기초

KOSPI200, KOSDAQ150, 미국S&P 등


- 업종/섹터추적 ETF

특정 업종 및 산업을 기초

금융, 반도체, 헬스케어 등


- 상품추적 ETF

금속, 곡물 등 상품을 기초

금, 은 콩


- 파생 ETF

기초자산을 2배 또는 역으로 추적하는 ETF

레버리지(+2배), 인버스(-1배)

10. ETN (Exchange Traded Note)

채권처럼 생긴 투자상품인데, 수익률은 ETF처럼 특정 지수를 따라감.

발행 금융기관의 신용 리스크가 존재 (ETF보다 위험할 수 있음).

(예) "신한 레버리지 WTI원유 선물 ETN"

11. IPO (Initial Public Offering)

기업이 처음으로 주식을 일반인에게 공개하고 상장하는 것.

상장 전에 공모가를 정하고 일반 투자자들이 청약을 신청함.

(예) 카카오뱅크, 두나무, 크래프톤 같은 IPO.

12. 관리종목

기업에 문제가 생겨 거래소가 지정한 주식.

사유: 감사의견 거절, 자본잠식, 사업보고서 미제출 등.

주가가 급락하거나 상장폐지 위험이 커짐.

13. 거래정지종목

일정 사유로 인해 주식 거래가 일시적으로 중단된 종목.

예: 급격한 주가 변동, 중요 공시 누락, 상장폐지 사유 발생.

14. 시장경보종목

이상 급등/급락하거나 거래량이 비정상적으로 늘어나는 종목에 대해 거래소가 주의/경고/위험 단계를 부여.

투자자에게 과열 리스크를 알리기 위해 경보.

15. 갭상승 (Gap Up)

주식이 전날 종가보다 훨씬 높은 가격으로 시작하는 현상.

주로 좋은 뉴스(실적 호조, 신제품 출시, 호재 등)로 발생.

16. 이격도과열 (Moving Average Convergence Divergence Overbought)

주가가 이동평균선 대비 너무 많이 올라서 과열 상태임을 나타냄.

이격도(乖離度)가 높으면 "조정 위험"이 커진다고 해석.

17. 투심과열 (Investor Sentiment Overheating)

투자자들의 심리가 너무 긍정적으로 쏠린 상태.

시장 전체가 "지나치게 낙관"하는 국면 (거품 위험 있음).

18. 상대강도과열 (Relative Strength Index Overbought)

RSI(상대강도지수)가 70~80 이상으로 과열되었음을 의미.

주식이 너무 많이 올라 과매수(overbought) 상태라고 판단.

19. 공매도거래현황

투자자들이 공매도를 얼마나 했는지 나타내는 통계.

공매도란, 주식을 빌려서 먼저 팔고 나중에 싸게 사서 갚는 투자법.

공매도가 많으면 주가 하락 압력이 커질 수 있음.

공매도의 핵심 개념
"비쌀 때 팔고, 쌀 때 산다" → 가격 하락에 베팅하는 전략



🔁 공매도 거래 구조

주식을 빌린다 (증권사나 기관 등에서)

현재 시장가에 판다

나중에 주가가 떨어지면 싸게 다시 사서

빌린 주식을 갚고 차익을 수익으로 얻는다

📉 예시

| 단계 | 설명                                    |
| -- | ------------------------------------- |
| ①  | 삼성전자 주가가 70,000원일 때 주식을 1주 빌려서 판다     |
| ②  | 주가가 60,000원으로 하락                      |
| ③  | 시장에서 60,000원에 다시 사서 빌린 주식 1주를 갚는다     |
| ④  | 차익 = 70,000 - 60,000 = **10,000원 이익** |

20. 시가총액 (Market Capitalization)

기업 전체의 가치 = 주가 × 발행주식 수.

시총이 크면 시장 영향력도 큼.

(예) 2025년 기준 삼성전자 시가총액은 약 500조 원 이상.

21. 다우산업 (Dow Jones Industrial Average)

미국 증시를 대표하는 30개 대형주로 구성된 주가지수.

오래된 전통 지수지만, 단순 평균 방식(주가 단순 합산)이어서 약간 단점도 있음.

22. 나스닥종합 (NASDAQ Composite)

미국 나스닥 시장 전체 상장 종목의 지수.

IT, 바이오, 테크 중심 기업들이 많음 (애플, 아마존, 테슬라 등).

변동성이 크고 혁신 성장주 투자 지표로 쓰임.

23. S&P 500 (Standard & Poor's 500)

미국 증시의 대표 지수.

미국 상위 500개 기업으로 구성.

시가총액 가중 평균 방식 → 시장 전체 흐름을 가장 잘 반영.

24. 니케이225 (Nikkei 225)

일본 도쿄증권거래소(TSE)의 대표 지수.

일본을 대표하는 225개 기업 포함.

유니클로(패스트리테일링), 소니, 도요타 등.

25. 상해종합 (Shanghai Composite Index)

중국 상하이 증권거래소의 종합 주가지수.

국영기업과 대형 중국기업 비중이 큼.

26. 항셍 (Hang Seng Index)

## 29. PER (Price to Earnings Ratio, 주가수익비율)
✔️ 정의
PER = 주가 ÷ 주당순이익 (EPS)

즉, 회사가 벌어들이는 이익에 비해 주가가 얼마나 비싼지를 나타냄

PER이 높다는 건 주가가 이익에 비해 비싸다, PER이 낮으면 저평가 가능성

✔️ 예시
어떤 회사의 주가가 50, 주당순이익(EPS)이 5라면
→ PER = 50 / 5 = 10

✔️ 해석
PER 10이면: 투자자가 이 회사의 이익을 바탕으로 10년 후에 원금을 회수할 수 있다는 의미 (단순 해석 시)

업종마다 적정 PER이 다르므로 비교 대상과 함께 보는 것이 중요

## 30. PBR (Price to Book Ratio, 주가순자산비율)
✔️ 정의
PBR = 주가 ÷ 주당순자산가치 (BPS)
또는
PBR = 시가총액 ÷ 자본총계

회사의 순자산에 비해 주가가 얼마나 비싼지/싼지를 평가

✔️ 예시
주가가 100원이고, 주당 순자산(BPS)이 50원이면
→ PBR = 100 / 50 = 2.0

✔️ 해석
PBR 1.0 미만: 시장가치가 순자산보다 낮다 → 저평가 가능성

PBR 2.0 이상: 시장이 자산보다 더 높은 가치를 매김 (브랜드력, 성장 기대 등 반영)



### PER vs PBR 비교 요약

| 항목   | PER         | PBR         |
| ---- | ----------- | ----------- |
| 기준   | 수익          | 자산          |
| 공식   | 주가 / EPS    | 주가 / BPS    |
| 의미   | 이익 대비 주가 수준 | 자산 대비 주가 수준 |
| 높을수록 | 고평가 가능성 ↑   | 고평가 가능성 ↑   |
| 낮을수록 | 저평가 가능성 ↑   | 저평가 가능성 ↑   |

```python
plt.figure(figsize=(14,6))
plt.plot(apple['Close'], label='Apple Close')
plt.plot(apple['SMA20'], label='Apple 20-Day SMA')
plt.title('Apple Close Price with 20-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
