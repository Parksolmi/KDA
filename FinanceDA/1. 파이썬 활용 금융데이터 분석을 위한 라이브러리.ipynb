### yfinance (Yahoo Finance API)
- 미국 주식은 심볼만, 한국 주식은 종목 코드 + 거래소 접미어를 붙여야 함.
  - .KS → KOSPI (유가증권시장) 종목, 예: 삼성전자 005930.KS
  - .KQ → KOSDAQ 종목, 예: 셀트리온헬스케어 091990.KQ

```python
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', '005930.KS', '000660.KS'] #애플 , 마이크로소프트, 구글, 아마존, 삼성전자, SK하이닉스 종목코드
prices = yf.download(tickers, start="2025-01-01", end="2025-05-06")['Close']
returns = prices.pct_change().dropna()
print('------prices-------')
print(prices)
print('------returns------')
print(returns)
```

```python
plt.figure(figsize=(8,6))
# plt.figure(figsize=(8,6), dpi=100) #800 x 600 사이즈 기본단위 인치치
sns.heatmap(returns.corr(), annot=True, cmap="coolwarm")
plt.title("Stock Correlation Heatmap")
plt.show()
```

### 코드 설명

plt.figure(figsize=(8, 6))

plt.figure()	새 그림(Figure) 객체 생성 – 그래프를 그릴 캔버스를 준비
figsize=(8, 6)	가로 8인치, 세로 6인치 크기로 그래프 캔버스를 설정

단위: "인치(inch)"
Matplotlib은 인치를 기본 단위로 씁니다.

DPI (Dot Per Inch) 기본값은 보통 100입니다.

따라서 8 x 6 인치이면 실제 출력 이미지 크기는 800 x 600 픽셀

### 코드 구성 상세 설명

#### 1. returns.corr()
returns는 보통 여러 자산의 수익률을 담은 Pandas DataFrame입니다.
예: 각 열이 종목이고, 각 행이 날짜인 형태.

returns.corr()는 자산 간의 상관계수 행렬 (correlation matrix) 를 반환합니다.

🧠 상관계수(Correlation):

+1: 완벽한 양의 상관관계 (같이 움직임)

0: 상관 없음

-1: 완벽한 음의 상관관계 (반대로 움직임)

#### 2. sns.heatmap(...)
Seaborn의 heatmap 함수는 2차원 행렬을 색깔로 표현한 시각화입니다.

여기에 상관관계 행렬을 넣으면, 각 자산 간의 상관성을 색으로 보여줍니다.

주요 옵션:
annot=True: 각 셀에 숫자(상관계수 값)를 표시.

cmap="coolwarm": 색상 맵 설정.

파란색(음의 상관관계), 흰색(0), 빨간색(양의 상관관계)

📈 예시 그림에서 기대할 수 있는 것

|      | AAPL | MSFT | TSLA |
| ---- | ---- | ---- | ---- |
| AAPL | 1.00 | 0.85 | 0.45 |
| MSFT | 0.85 | 1.00 | 0.40 |
| TSLA | 0.45 | 0.40 | 1.00 |


대각선은 항상 1.00 (자기 자신과의 상관관계)

AAPL과 MSFT는 매우 높은 상관관계 (0.85, 진한 빨간색)

TSLA는 상대적으로 낮은 상관관계 (0.4~0.45, 연한색)

### 누적 수익률 그래프
- 1을 기준으로 1 이하면 적자, 1 이상이면 흑자

```python
cumulative = (1 + returns).cumprod()
cumulative.plot(figsize=(12,6), title="Cumulative Returns")
plt.ylabel("Growth of $1")
plt.grid(True)
plt.show()
```

### 코드 설명
1. (1 + returns)  
returns는 일반적으로 일간/주간/월간 수익률 시리즈 (예: Pandas Series)입니다.

1 + returns는 각 기간의 수익률을 1에 더한 값입니다.  
예: 수익률이 5%라면 → 1 + 0.05 = 1.05 (즉, 원금이 5% 증가했다는 의미)

2. .cumprod()  
누적 곱 (cumulative product)을 계산합니다.

예를 들어, 연속된 수익률이 [1.05, 0.97, 1.02]라면, 누적 곱은:

[1.05,  
 1.05 * 0.97 = 1.0185,  
 1.0185 * 1.02 = 1.03987]  
 
이것은 원금이 시간에 따라 어떻게 변화했는지를 보여줍니다.

🟡 결과적으로 cumulative은 투자 초기 금액이 1이라고 가정했을 때, 그 금액이 시간이 지남에 따라 어떻게 변했는지를 나타냅니다.

### twinx()를 사용한 이중 y축 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

# 샘플 데이터
x = np.linspace(0, 10, 100)
y1 = np.sin(x)               # 첫 번째 데이터 (작은 값)
# y2 = 100 * np.exp(-0.3 * x)  # 두 번째 데이터 (큰 값)
y2 = 2*x*x + 10

# 첫 번째 y축
fig, ax1 = plt.subplots()

ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_ylabel('sin(x)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 두 번째 y축 (공유 x축)
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r--', label='exp decay')
# ax2.set_ylabel('100 * exp(-0.3x)', color='r')
ax2.set_ylabel('2 * x * x + 10', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title("Two Garph overlapped")
plt.show()
```

**코드설명**

tick_params(...): 눈금과 관련된 설정을 변경할 때 사용하는 함수입니다.

axis='y': y축에만 이 설정을 적용한다는 의미입니다.

labelcolor='r': 눈금 라벨(숫자 등)의 색상을 빨간색('r')으로 설정합니다.

요약
ax1.twinx()를 사용하면 같은 x축을 공유하면서 y축을 두 개 사용할 수 있습니다.

각 y축에 다른 색상과 라벨을 줘서 시각적으로 구분이 쉽게 합니다.

### 5. 이동 평균 (SMA, EMA)
