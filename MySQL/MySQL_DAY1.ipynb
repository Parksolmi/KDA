```python
# 덩어리 다 가져오기
div_list = news_bs.find_all('div', {'class': 'item-bundle-mid'})

for div in div_list[ :10]:
    print(div.find('a').text)
    print(div.find('a')['href'])
    print(div.find('p').text)
    print(' ')
```

```python
import pandas as pd

pd.DataFrame(new_list)
```

```python
# URL 셋팅
def get_newsData(search, page):
    url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=STC&enc=utf8&cluster=y&cluster_page=1&q={}&p={}&sort=recency'
    real_url = url.format('search', 3)

    resp = requests.get(real_url)
    news_bs = bs4.BeautifulSoup(resp.text)

    div_list = news_bs.find_all('div', {'class': 'item-bundle-mid'})[:10]
    news_list = []
    for div in div_list:
        temp = {}
        temp['title'] = div.find('a').text
        temp['link'] = div.find('a')['href']
        temp['summary'] = div.find('p').text
        news_list.append(temp)

    return news_list
```

```python
news_result = []
for p in range(1,11):
    news_result += get_newsData('경제', p)
```

```python
result = pd.DataFrame(news_result)
display(result)
```
