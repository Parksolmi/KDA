```python
#9ë²ˆ

# í•™ìƒë“¤ì˜ ì„±ì  ì •ë³´ë¥¼ ë‹´ì€ ë”•ì…”ë„ˆë¦¬ì…ë‹ˆë‹¤. ì´ ì½”ë“œëŠ” ìˆ˜ì •í•˜ì§€ ë§ˆì„¸ìš”.
student_scores = {
    "Peter": 85,
    "Sarah": 92,
    "John": 78
}

# 1. Sarahê°€ ìê¸° ì ìˆ˜ë¥¼ ê¶ê¸ˆí•´ í•©ë‹ˆë‹¤. ë”•ì…”ë„ˆë¦¬ student_scoresë¥¼ ì‚¬ìš©í•´ Sarahì˜ ì ìˆ˜ë¥¼ ì¶œë ¥í•´ë³´ì„¸ìš”.
print(f"Sarahì˜ ì ìˆ˜: {student_scores['Sarah']}")

# 2. ìƒˆë¡œìš´ í•™ìƒ Mayê°€ ì…í•™í•˜ê³  ì‹œí—˜ì„ ì³¤ìŠµë‹ˆë‹¤. ë”•ì…”ë„ˆë¦¬ student_scoresì— ìƒˆë¡œìš´ í•™ìƒê³¼ í•™ìƒì˜ ì ìˆ˜ë¥¼ ì¶”ê°€í•˜ê³ , ë”•ì…”ë„ˆë¦¬ë¥¼ ì¶œë ¥í•´ë³´ì„¸ìš”.
student_scores["May"] = 99
print(f"Mayê°€ ì…í•™í•œ í›„ ë”•ì…”ë„ˆë¦¬: {student_scores}")

# 3. ì±„ì  ê³¼ì •ì— ì‹¤ìˆ˜ê°€ ìˆì—ˆë„¤ìš”. ë”•ì…”ë„ˆë¦¬ì—ì„œ student_scoresì—ì„œ Johnì˜ ì ìˆ˜ë¥¼ ë‹¤ë¥¸ ì ìˆ˜ë¡œ ë°”ê¾¸ì–´ ì£¼ì„¸ìš”.
student_scores["John"] = 98
print(f"Johnì˜ ì ìˆ˜ë¥¼ ìˆ˜ì •í•œ í›„ ë”•ì…”ë„ˆë¦¬: {student_scores}")

# 4. Sarahê°€ ëŒì—° í‡´í•™í–ˆìŠµë‹ˆë‹¤. ë”•ì…”ë„ˆë¦¬ì—ì„œ Sarahì˜ ì •ë³´ë¥¼ ì‚­ì œí•´ì£¼ì„¸ìš”. ë”•ì…”ë„ˆë¦¬ì—ì„œ ì•„ì´í…œì„ ì‚­ì œí•˜ë ¤ë©´ del ë”•ì…”ë„ˆë¦¬ëª…[í‚¤]ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
del student_scores["Sarah"]
print(f"Sarahê°€ í‡´í•™í•œ í›„ ë”•ì…”ë„ˆë¦¬: {student_scores}")

# 5. ìš°ë¦¬ í•™êµì— Jakeê°€ ìˆì—ˆë˜ê°€ìš”? ìš°ë¦¬ í•™êµì— Peterê°€ ìˆì—ˆë˜ê°€ìš”? í™•ì¸í•´ì£¼ì„¸ìš”.
print(f"is Jake in dict? {'Jake' in student_scores}")
print(f"is Peter in dict? {'Peter' in student_scores}")

# 6. ì•„ë¬´ë˜ë„ ëª¨ë“  í•™ìƒë“¤ì˜ ì´ë¦„ì„ í™•ì¸í•´ì•¼ê² ìŠµë‹ˆë‹¤. ë”•ì…”ë„ˆë¦¬ì˜ ëª¨ë“  keyë“¤ì„ í™•ì¸í•˜ë ¤ë©´ dictionaryëª….keys()ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
print(f"ëª¨ë“  í•™ìƒë“¤ì˜ ì´ë¦„: {student_scores.keys()}")

# 7. ì´ì   ëª¨ë“  í•™ìƒë“¤ì˜ ì ìˆ˜ë¥¼ í™•ì¸í•´ì•¼ê² ìŠµë‹ˆë‹¤. ë”•ì…”ë„ˆë¦¬ì˜ ëª¨ë“  valueë“¤ì„ í™•ì¸í•˜ë ¤ë©´ dictionaryëª….values()ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
print(f"ëª¨ë“  í•™ìƒë“¤ì˜ ì ìˆ˜: {student_scores.values()}")

# 8. ìš°ë¦¬ í•™êµëŠ” ìµœì •ì˜ˆ í•™ìƒë“¤ë¡œ êµ¬ì„±ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ì´ ëª‡ ëª…ì´ ìˆì„ê¹Œìš”?
print(f"ì´ í•™ìƒ ìˆ˜: {len(student_scores)}")
```
