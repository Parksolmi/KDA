````python
A = True
B = False
C = True

# 1번
ex1 = A or B
ans1 = None

# 2번
ex2 = A and not B
ans2 = None

# 3번
ex3 = B or not C
ans3 = None

# 3번
ex4 = not A or B and not C
ans4 = None

# 결과를 출력하겠습니다.
print(f"1번 답변: {ans1}, 실제 정답: {ex1}")
print(f"2번 답변: {ans2}, 실제 정답: {ex2}")
print(f"3번 답변: {ans3}, 실제 정답: {ex3}")
print(f"4번 답변: {ans4}, 실제 정답: {ex4}")
````

청소기는 먼지를 감지하면 `1`씩 카운트를 증가시키고,  
**총 15개의 먼지를 청소하면 작동을 멈춰야** 합니다.  

하지만 지금의 청소기는 **무한히 먼지를 센 뒤 계속 작동 중**이에요!  
`while True:`는 수정하지 않고, 조건에 따라 루프를 종료하도록 코드를 고쳐봅시다.


### 🍎 **출력 예시**
```
청소한 먼지 개수: 15
청소 완료!
```

### 🤔 **지시사항**
1. 아래 코드를 확인하고 **무한 루프가 발생하는 이유를 파악**해보세요.
2. `break` 문을 사용해 먼지 15개를 청소하면 루프를 탈출하도록 코드를 수정하세요.
